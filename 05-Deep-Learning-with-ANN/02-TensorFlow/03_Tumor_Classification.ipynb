## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

## Reading Dataset

```python
df = pd.read_csv('cancer_classification.csv')
```

```python
df.info()
```

```python
df.describe().transpose()
```

## Exploratory Data Analysis

```python
sns.countplot(x='benign_0__mal_1', data=df);
```

```python
df.corr()['benign_0__mal_1'][:-1].sort_values().plot(kind='bar');
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr());
```

## Data Preprocessing

```python
X = df.drop('benign_0__mal_1', axis=1).values
y = df['benign_0__mal_1'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=101
)
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train.shape
```

## Creating Model

```python
model = Sequential()

model.add(Dense(units=30, activation='relu'))
model.add(Dense(units=15, activation='relu'))
model.add(Dense(units=1, activation='sigmoid'))

model.compile(
    optimizer='adam',
    loss='binary_crossentropy'
)
```

## Training the Model

```python
model.fit(x=X_train, y=y_train, validation_data=(X_test, y_test), epochs=600)
```

## Model Evaluation

```python
losses_df = pd.DataFrame(model.history.history)
```

```python
losses_df.plot();
```

## Recreating and Training the Model with Early Stopping

```python
early_stop = EarlyStopping(monitor='val_loss',
                           mode='min', verbose=1,
                           patience=25)
```

```python
model = Sequential()

model.add(Dense(units=30, activation='relu'))
model.add(Dense(units=15, activation='relu'))
model.add(Dense(units=1, activation='sigmoid'))

model.compile(
    optimizer='adam',
    loss='binary_crossentropy'
)
```

```python
model.fit(x=X_train, y=y_train, validation_data=(X_test, y_test),
          epochs=600, callbacks=[early_stop])
```

```python
losses_df = pd.DataFrame(model.history.history)
losses_df.plot();
```

## Recreating and Training the Model with Dropout Layer

```python
model = Sequential()

model.add(Dense(units=30, activation='relu'))
model.add(Dropout(rate=0.5))

model.add(Dense(units=15, activation='relu'))
model.add(Dropout(rate=0.5))

model.add(Dense(units=1, activation='sigmoid'))

model.compile(
    optimizer='adam',
    loss='binary_crossentropy'
)
```

```python
model.fit(x=X_train, y=y_train, validation_data=(X_test, y_test),
          epochs=600, callbacks=[early_stop])
```

```python
losses_df = pd.DataFrame(model.history.history)
losses_df.plot();
```

## Prediction

```python
predictions = (model.predict(X_test) > 0.5).astype(int)
```

```python
print(classification_report(y_true=y_test, y_pred=predictions))
```

```python
plt.figure(figsize=(4,3))
matrix = confusion_matrix(y_true=y_test, y_pred=predictions)
sns.heatmap(matrix, annot=True, fmt='1.0f');
```
